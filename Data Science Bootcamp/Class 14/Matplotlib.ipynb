```python
import numpy as np
```

```python
x = np.arange(1,11)
y = np.arange(11,21)
```

```python
plt.scatter(x,y , c = "g")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Numpy Arrays")
plt.savefig("hello.png")
```

```python
plt.plot(x , y , "g*", linestyle = "dashed" , linewidth = 2 , markersize = 20)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,c="g")
plt.subplot(2,2,2)
plt.plot(x,y,"r--")
plt.subplot(2,2,3)
plt.plot(x,y,"go")
plt.subplot(2,2,4)
plt.plot(x,y,"bo")         
```

```python
x = np.arange(0,4*np.pi)
y = np.sin(x)
plt.title("Sine Wave")
plt.plot(x,y)
```

```python
x = np.arange(0,5*np.pi)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x , y_sin , "g--")
plt.title("Sine")

plt.subplot(2,1,2)
plt.plot(x , y_cos , "b--")
plt.title("Cos")
```

```python
x = [2,4,8]
y = [3,5,7]

x2 = [4, 9, 14]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2 , y2 , color = "g")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar Graph")
```

```python
a = np.array([4,6,8,9,12,23,45,89])
plt.hist(a)
plt.title("Histogram")
```

```python
data = [np.random.normal(0 , std , 100) for std in range(1, 4)]
plt.boxplot(data)
```
